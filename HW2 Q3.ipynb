```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics
```

```python
likelihood_w1 = np.arange(5, 35, 0.01)
likelihood_w2 = np.arange(5, 35, 0.01)
plt.plot(likelihood_w1, norm.pdf(likelihood_w1, 20, 4))
plt.plot(likelihood_w2, norm.pdf(likelihood_w2, 15, 2))
plt.show()
```

```python
# After normalizing (Posterior)
prior_w1 = 1/3
prior_w2 = 2/3
post_w1 = norm.pdf(likelihood_w1, 20, 4) * prior_w1
post_w2 = norm.pdf(likelihood_w2, 15, 2) * prior_w2
plt.plot(likelihood_w1, post_w1)
plt.plot(likelihood_w2, post_w2)

plt.scatter(17,norm.pdf(17, 20, 4) * prior_w1)
plt.scatter(17, norm.pdf(17, 15, 2) * prior_w2)

idx = np.argwhere(np.diff(np.sign((norm.pdf(likelihood_w1, 20, 4) * prior_w1) - norm.pdf(likelihood_w2, 15, 2) * prior_w2))).flatten()
plt.plot(likelihood_w1[idx], post_w1[idx], 'ro')

plt.show()
```

```python
print(norm.pdf(17, 20, 4) * prior_w1 < 17, norm.pdf(17, 15, 2) * prior_w2)
```
